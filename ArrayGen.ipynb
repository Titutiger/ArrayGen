`.expr()`

This is a function used to get co-domains from iterated domains.
Simply, f(x) = y where x can be iterated from ... -> ... and y is the output of f(x).

This is how it goes:

````python
x, y = q.expr('3*x**2 + 5*x + 4', vars_='x')

print(f'Domains: {x}')
print(f'Co-domains: {y}')
````

Herein, the equation is:

$ f(x) = 3x^2 + 5x + 4 $

Wherein:

1st iteration: ---> $ x = 1 $ therefore $ f(1) = 12 = y $

2nd iteration: ---> $ x = 2 $ therefore $ f(2) = 26 = y $

and so on ... ...

The `f(x)` values will be taken as `y` and they will be plotted on the y-axis, which brings me to the next part:

With `.expr()`, the primary thing that one can do for now is to visualise what this equation does.
With the above equation, we will make a graph. To do this, use `.Graphing.plot()`:

````python
q.Graphing.plot(x, y, static=True)
````

But, a more interesting way to use this is with 3 variables:

Now, let's try something similar:

Herein, the formula will be:

$ f(x) = x^2 + 3xy + y^3$

````python
x, y, z = q.expr('x**2 + 3*x*y + y**3', vars_="x y")
q.Graphing.plot(x, y, z=z, static=True)
````

Themes and colors can also be applied:

````python
q.Graphing.plot(x, y, z=z, static=True, color='twilight')
````

`.Maths.permutations()` uses cnr,

$ \binom{n}{r} = \frac{n!}{r!(n-r)!} $

Wherein 10 is `n` and 5 is `r` in the given example:

````python
# Permutations:

x = q.Maths.permutations(10, 5)
x
````

`.Maths.Complex.z_algebra` is really straight forward.

It separates the real and imaginary components and performs the given operation.

$ x = 2 + 3i , y = 6 + 7i $

$ x + y = 2 + 3i + 6 + 7i $

$= 2 + 6 + 3i + 7i $

$= 8 + 10i $

Out of which, it returns only the integers/floats and not iota.

````python
# Complex:

x = q.Maths.Complex.z_algebra('2+3i', '6+7i', operation='+')
print(x)

# or

real, imag = q.Maths.Complex.z_algebra('2+3i', '6+7i', operation='-')
print(real, imag)
````

````python
# Complex:

x = q.Maths.Complex.z_forms('2+3i')
print(x)

y = q.Maths.Complex.z_forms('2+3i', values=True)
print(y)
````

````python
# Hypo:

x = q.Maths.hypo(3.0, 4.0, 90.0)
x
````

````python
# Progressions:

x = q.Maths.Progressions.term('ap', '2, 4, 6, 8', 10)
x
````

````python
x = q.Maths.Progressions.term('gp', '2 4 8 16', 5)
x
````

````python
# Arithmetic Progression (AP): first term = 2, difference = 3, 5 terms
ap_sequence = q.Maths.Progressions.get_progression('ap', 2, d=3, n=5)
# Output: [2.0, 5.0, 8.0, 11.0, 14.0]
print(ap_sequence)
# Geometric Progression (GP): first term = 3, ratio = 2, 4 terms
gp_sequence = q.Maths.Progressions.get_progression('gp', 3, r=2, n=4)
# Output: [3.0, 6.0, 12.0, 24.0]
print(gp_sequence)
# Harmonic Progression (HP): first term = 1, difference = 1, 4 terms
hp_sequence = q.Maths.Progressions.get_progression('hp', 1, d=1, n=4)
# Output: [1.0, 0.5, 0.3333333333333333, 0.25]
print(hp_sequence)
# Arithmetic-Geometric Progression (AGP): first term = 1, difference = 2, ratio = 3, 4 terms
agp_sequence = q.Maths.Progressions.get_progression('agp', 1, d=2, r=3, n=4)
# Output: [1.0, 9.0, 30.0, 81.0]
print(agp_sequence)
# To get usage help
help_text = q.Maths.Progressions.get_progression('help')
print(help_text)
````

````python
# Mandelbrot Sets:
"""Using `.Graphing`"""

# 1. Classic Mandelbrot: z = z^2 + c
formula_square = lambda z, c: z*z + c
m = q.Maths.mandelbrot_set(formula_square)
q.Graphing.mandelbrot(m)

# 2. Cubic Mandelbrot: z = z^3 + c
formula_cube = lambda z, c: z**3 + c
m = q.Maths.mandelbrot_set(formula_cube)
q.Graphing.mandelbrot(m)

# 3. Quartic Mandelbrot: z = z^4 + c
formula_quartic = lambda z, c: z**4 + c
m = q.Maths.mandelbrot_set(formula_quartic)
q.Graphing.mandelbrot(m)

# 4. Mandel bar (Tricorn): z = conjugate(z)^2 + c
formula_mandelbar = lambda z, c: (z.conjugate())**2 + c
m = q.Maths.mandelbrot_set(formula_mandelbar)
q.Graphing.mandelbrot(m)

# 5. Burning Ship fractal: z = (|Re(z)| + i|Im(z)|)^2 + c
formula_burning_ship = lambda z, c: complex(abs(z.real), abs(z.imag))**2 + c
m = q.Maths.mandelbrot_set(formula_burning_ship)
q.Graphing.mandelbrot(m)
````
